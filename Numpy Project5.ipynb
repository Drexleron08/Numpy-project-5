```python
#to verify the monte carlo pi method 
import numpy as np
#function to provide us with a data of random points in a unit square
def create_random_points(n):
    points = np.random.uniform(-1, 1, (n, 2))
    return points
```

```python
#to find the distance of the points from the origin and check if they are inside the unit circle or not and then estimate the value of pi
def estimate_pi():
    n=99999999
    points = create_random_points(n)
    distance=np.sum(points**2, axis=1)
    inside_circle = np.sum(distance <= 1)
    pi_estimate = 4*(inside_circle / n) 
    print(f"Total number of points generated: {n}")
    print(f"Number of points inside the circle: {inside_circle}")
    print(f"Estimated value of pi: {pi_estimate}")
estimate_pi()

    
```
